```python
def max_value(nodes, i, alpha, beta):
    nodes[i]['visited'] = True
    if len(nodes[i]['children']) == 0:
        return nodes[i]['v']
    else:
        v = -float('inf')
        nodes[i]['alpha'] = alpha
        nodes[i]['beta'] = beta
        for child in nodes[i]['children']:
            v  = max(v, min_value(nodes, child, nodes[i]['alpha'], nodes[i]['beta']))
            if v >= nodes[i]['beta']:
                nodes[i]['v'] = v
                nodes[i]['alpha'] = max(v, nodes[i]['alpha'])
                return v
            nodes[i]['alpha'] = max(v, nodes[i]['alpha'])
        nodes[i]['v'] = v
        return v
```

```python
def min_value(nodes, i, alpha, beta):
    nodes[i]['visited'] = True
    if len(nodes[i]['children']) == 0:
        return nodes[i]['v']
    else:
        v = float('inf')
        nodes[i]['alpha'] = alpha
        nodes[i]['beta'] = beta
        for child in nodes[i]['children']:
            v = min(v, max_value(nodes, child, nodes[i]['alpha'], nodes[i]['beta']))
            if v <= nodes[i]['alpha']:
                nodes[i]['v'] = v
                nodes[i]['beta'] = min(v, nodes[i]['beta'])
                return v
            nodes[i]['beta'] = min(v, nodes[i]['beta'])
        nodes[i]['v'] = v
        return v
```
